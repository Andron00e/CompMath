```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

# Задача 5

```python
dataset = sns.load_dataset('anscombe')
data = []
data.append((dataset[dataset['dataset'] == "I"]["x"].to_numpy(), dataset[dataset['dataset'] == "I"]["y"].to_numpy()))
data.append((dataset[dataset['dataset'] == "II"]["x"].to_numpy(), dataset[dataset['dataset'] == "II"]["y"].to_numpy()))
data.append((dataset[dataset['dataset'] == "III"]["x"].to_numpy(), dataset[dataset['dataset'] == "III"]["y"].to_numpy()))
data.append((dataset[dataset['dataset'] == "IV"]["x"].to_numpy(), dataset[dataset['dataset'] == "IV"]["y"].to_numpy()))
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
for i in range(4):
    x = data[i][0]
    y = data[i][1]
    axs[int(i/2)][i%2].set_title("Dx={:.0f}, Dy={:.2f}, r={:.2f}".format(np.var(x), np.var(y), np.corrcoef(x, y)[0,1]))
    axs[int(i/2)][i%2].scatter(x, y, c="b")
```

# Задача 7

```python
true_w1 = 2.8
true_w0 = 1.5
xs = np.arange(0, 10, 0.5)
noise = np.random.normal(0, 0.45, size=xs.shape)
ys = true_w1 * xs + true_w0 + noise
w1 = np.cov([xs,ys],bias=True)[0][1]/np.var(xs)
w0 = ys.mean() - w1 * xs.mean()
print("w0=",w0,", w1=",w1)
plt.scatter(xs,ys)
plt.plot(xs, w0+w1*xs)
plt.show()
```
